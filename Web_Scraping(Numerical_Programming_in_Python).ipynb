```python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
```

## **Scrapping Movies Data**

```python
# Specifying the URL from which movies related data will be fetched

# WAS GETTING 403 ERROR, so took help online and found this piece of code so that it wont throw 403 error
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
}
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url, headers= headers)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
print(soup.prettify())
```

## **Fetching Movie URL's**

```python
movie_url = []
for x in soup.find_all('a',attrs={'class':'title-list-grid__item--link'}):
  movie_url.append('https://www.justwatch.com'+x['href'])
print(movie_url)
print(len(movie_url))
```

## **Scrapping Movie Title**

```python
# Write Your Code here
#import time
movie_title = []
for url in movie_url:
  try:
    headers =headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    title = soup.find_all('div',attrs={'class':'title-detail-hero__details'})[0].find_all('h1')[0].text.strip()
  except:
    title = "NA"
  movie_title.append(title)
  #time.sleep(2)
print(movie_title)
print(len(movie_title))
```

## **Scrapping release Year**

```python
#import time
release_year = []
for url in movie_url:
  try:
    headers = headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    year = eval(soup.find_all('div',attrs={'class':'title-detail-hero__details'})[0].find_all('span')[0].text.strip())
  except:
    year = "NA"
  release_year.append(year)
  #time.sleep(3)
print(release_year)
print(len(release_year))
```

## **Scrapping Genres**

```python
#import time
movie_genres = []
for url in movie_url:
  try:
    headers =headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    for x in soup.find_all('div',attrs={'class':'detail-infos'}):
      if x.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text=='Genres':
        genres = x.find_all('div',attrs={'class':'detail-infos__value'})[0].text

  except:
    genres = "NA"
  movie_genres.append(genres)
  #time.sleep(2)
print(movie_genres)
print(len(movie_genres))
```

## **Scrapping IMBD Rating**

```python
# Write Your Code here
#import time
imdb_rating = []
for url in movie_url:
  try:
    headers = headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    for x in soup.find_all('div',attrs={'class':'detail-infos'}):
      if x.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text=='Rating':
        rating = x.find_all('div',attrs={'class':'detail-infos__value'})[0].text.strip()
  except:
    rating = 'NA'
  imdb_rating.append(rating)
  #time.sleep(3)
print(imdb_rating)
print(len(imdb_rating))
```

## **Scrapping Runtime/Duration**

```python
runtime = []
for url in movie_url:
  try:
    headers = headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    for x in soup.find_all('div',attrs={'class':'detail-infos'}):
      if x.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text=='Runtime':
        rn_time = x.find_all('div',attrs={'class':'detail-infos__value'})[0].text
  except:
    runtime = 'NA'
  runtime.append(rn_time)

print(runtime)
print(len(runtime))
```

## **Scrapping Age Rating**

```python
age_rating = []
for url in movie_url:
  try:
    headers = headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    age = 'NA'
    for x in soup.find_all('div',attrs={'class':'detail-infos'}):
      if x.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text=='Age rating':
        age = x.find_all('div',attrs={'class':'detail-infos__value'})[0].text
    age_rating.append(age)
  except:
    age_rating.append('NA')

print(age_rating)
print(len(age_rating))
```

## **Fetching Production Countries Details**

```python
production_country = []
for url in movie_url:
  try:
    headers = headers
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,'html.parser')
    for x in soup.find_all('div',attrs={'class':'detail-infos'}):
      if x.find_all('h3',attrs={'class':'detail-infos__subheading'})[0].text==' Production country ':
         country = x.find_all('div',attrs={'class':'detail-infos__value'})[0].text
  except:
    country  = 'NA'
  production_country.append(country)

print(production_country)
print(len(production_country))
```

## **Fetching Streaming Service Details**

```python
# Write Your Code here
Movie_Streaming_Provider =[]
#import time
for url in movie_url:
  try:
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.9',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
}
    data = requests.get(url, headers = headers)
    soup = BeautifulSoup(data.text,"html.parser")

    Stream_provider = soup.find("div",attrs ={'class':'picture-wrapper'}).find_all("img",attrs={'class':'offer__icon'})
    alt_values = [img['alt'] for img in Stream_provider]
    alt_values = (",".join(alt_values))

  except :
    alt_values = 'NA'
  Movie_Streaming_Provider.append(alt_values)
  #time.sleep(3)


print(Movie_Streaming_Provider)
print(len(Movie_Streaming_Provider))
```

```python
import pandas as pd
import numpy as np

# Replace non-numeric values with NaN
JustWatch_df['Release_year'] = pd.to_numeric(JustWatch_df['Release_year'], errors='coerce')

# Drop rows with NaN values in Release_year
JustWatch_df = JustWatch_df.dropna(subset=['Release_year'])

# Convert Release_year to integer
JustWatch_df['Release_year'] = JustWatch_df['Release_year'].astype(int)

# Define filter criteria
start_date = 2022
end_date = 2024
min_rating = 7.0

# Filter the DataFrame
filter_data_df = JustWatch_df[
    (JustWatch_df['Release_year'] >= start_date) &
    (JustWatch_df['Release_year'] <= end_date) &
    (JustWatch_df['imdb_rating'] >= min_rating)
]


```
